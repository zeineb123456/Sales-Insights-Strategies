# Découverte des données 

```python
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#collecter les noms des fichiers (dataset)
files = [file for file in os.listdir(r'C:\Users\Asus\Desktop\Analyse des vente mensuelle\données')]
for file in files :
    print(file)
```

```python
path = r'C:\Users\Asus\Desktop\Analyse des vente mensuelle\données'

#créer une bdd vide
all_data = pd.DataFrame()

for file in files:
    current_data = pd.read_csv(path+'/'+file)
    all_data=pd.concat([all_data, current_data])
print(all_data)
```

```python
all_data.to_csv(path + 'all_data.csv', index = False)
```

```python
#supprimer les valeurs manquantes
all_data = all_data.dropna(how='all')
all_data.shape
```

# Quel est le mois durant lequel nous avons réalisé le meilleur chiffre d'affaire?

```python
def month(x):
    return x.split('/')[0]
```

```python
all_data['Month'] = all_data['Order Date'].apply(month)
all_data
```

```python
all_data['Month'].unique()
```

```python
all_data = all_data[all_data['Month']!= 'Order Date']
all_data['Month'].unique()
```

```python
all_data.dtypes
```

```python
all_data['Month'] = all_data['Month'].astype(int)
all_data.dtypes
```

```python
all_data['Price Each'] = all_data['Price Each'].astype(float)
all_data['Quantity Ordered'] = all_data['Quantity Ordered'].astype(int)
all_data.dtypes
```

```python
all_data['Sales'] = all_data['Quantity Ordered']*all_data['Price Each']
all_data
```

```python
all_data.groupby('Month') ['Sales'].sum()
```

```python
months = range(1,13)
plt.bar(months,all_data.groupby('Month') ['Sales'].sum())
plt.xticks(months)
plt.ylabel('Sales in USD')
plt.xlabel('Month number')
plt.title('Total Sales per Month (in USD)')
plt.savefig('total_sales_per_month.png')
plt.show()
```

# En quel moment doit-on faire une campagne publicitaire pour avoir plus de ventes ?

```python
all_data['Hour'] = pd.to_datetime(all_data['Order Date']).dt.hour
all_data
```

```python
keys = []
hours = []
for key, hour in all_data.groupby('Hour'):
    keys.append(key)
    hours.append(len(hour))
hours
```

```python
plt.grid()
plt.plot(keys, hours)
plt.xlabel('Hour of the Day') 
plt.ylabel('Number of Orders')  
plt.title('Number of Orders Throughout the Hours')  
plt.savefig('Number of Orders Throughout the Hours.png') 
plt.show()
```

# Quel produit se vend le plus ?

```python
import matplotlib.pyplot as plt

all_data.groupby('Product')['Quantity Ordered'].sum().plot(kind='bar')

plt.title('Produit le plus vendu')

plt.savefig('produit_le_plus_vendu.png')

plt.show()
```

```python
all_data.groupby('Product')['Price Each'].mean()
```

```python
products = all_data.groupby('Product')['Quantity Ordered'].sum().index
quantity = all_data.groupby('Product')['Quantity Ordered'].sum()
prices = all_data.groupby('Product')['Price Each'].mean()
```

```python
plt.figure(figsize=(40,24))
fig, ax1 = plt.subplots()
ax2 = ax1.twinx()

ax1.bar(products, quantity, color='g')
ax2.plot(products, prices, 'b-')

ax1.set_xticklabels(products, rotation='vertical', size=8)
ax1.set_xlabel('Products')
ax1.set_ylabel('Quantity Sold', color='g')
ax2.set_ylabel('Average Price', color='b')

plt.title('Sales Performance by Product')
plt.savefig('Sales Performance by Product.png')
plt.show()
```

# Quelles sont les combinaisons des produits qui se vendent le plus ?

```python
#chercher les combinaison
df=all_data[all_data['Order ID'].duplicated(keep=False)]
```

```python
df['Grouped'] = df.groupby('Order ID')['Product'].transform(lambda x: ','.join(x))
```

```python
#supprimer les elements dupliquer
df2 = df.drop_duplicates(subset=['Order ID'])
df2
```

```python
df2['Grouped'].value_counts()[0:5].plot.pie(title='Most Purchased Product Combinations')
plt.savefig('Most Purchased Product Combinations.png')
```
